#1. What is a random variable in probability theory?
 - In probability theory, a random variable is a numerical value assigned to the outcome of a random experiment. Simply put, it provides a way to map the results of an unpredictable event (like flipping a coin or rolling a dice) into numbers that we can analyze mathematically.

#2. What are the types of random variables?
 - There are two main types of random variables:

1. Discrete Random Variable: Takes on a countable set of values, such as the outcome of a dice roll (1, 2, 3, 4, 5, 6).

2. Continuous Random Variable: Can take on any value within a continuous range, like the temperature or the height of people in a population.

#3. What is the difference between discrete and continuous distributions?
 - Difference between **discrete** and **continous distributions**.

1. **Discrete Distributions:**                            
   - It deals with countable outcomes(whole numbers).
   - Probabilities are assigned to individual values. for example, the probability of rolling a specific number on a die.
   - Common examples: **Binomal distribution**, **Poisson distribution**.

2. **Continous Distributions:**
   - It deals with uncountable outcomes(real number in a range).
   - Probabilities are assigned to intervals, not individual values. for example, the probability of a person's height falling between 150 cm and 160 cm.
   - Common examples: **Normal distribution**, **Exponential distribution**.


#4. What are probability distribution functions (PDF)?
 - In probability theory, a probability distribution function (PDF) describes the likelihood of a random variable taking on specific values.

**For discrete random variables:**
The PDF is often referred to as a probability mass function (PMF). It assigns probabilities to individual values. For example, rolling a die: the PMF gives the probability of rolling any specific number (1 through 6).

**For continuous random variables:**
The PDF represents the probability density over a range of values. Instead of giving probabilities for exact values (which is 0 for continuous variables), it tells us how "dense" the probabilities are in a region. For example, in a normal distribution, the PDF is a bell-shaped curve.

#5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
 - Difference between cumulative distribution functions (CDF) and probability distribution functions (PDF).

   - For continuous variables, the PDF gives a curve; probabilities are found by calculating the area under the curve over a specific range.

  - The CDF shows the cumulative probability up to a particular value.

  - The CDF is a non-decreasing function and always ranges from 0 to 1.

#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculates the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): The probability of success (value of 1), 0 <= p <= 1.

    Returns:
    float: Probability of the outcome.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be in the range [0, 1].")

    return p if x == 1 else 1 - p

p_success = 0.7
outcome = 1
probability = bernoulli_pdf(outcome, p_success)
print(f"The probability of outcome {outcome} is {probability}.")
```

#9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
    data (list or numpy array): The dataset.

    Returns:
    numpy array: Z-scores of the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

data = np.random.normal(10, 2, 100)
z_scores = calculate_z_scores(data)

plt.scatter(range(len(z_scores)), z_scores, color='purple', alpha=0.7, label='Z-scores')
plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.xlabel('Data Index')
plt.ylabel('Z-score')
plt.title('Z-scores Visualization')
plt.legend()
plt.grid(alpha=0.6)
plt.show()
```

#12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).

```python
import numpy as np
import matplotlib.pyplot as plt

def standard_normal_distribution():
    """
    Calculates and plots the standard normal distribution (mean = 0, std = 1).
    """
    x = np.linspace(-4, 4, 500)
    mean = 0
    std = 1

    pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

    plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')
    plt.fill_between(x, pdf, color='skyblue', alpha=0.4)
    plt.xlabel('Value (Z)')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution (mean = 0, std = 1)')
    plt.grid(alpha=0.6)
    plt.legend()
    plt.show()

standard_normal_distribution()
```

#13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.

```python
import numpy as np

def binomial_distribution(n, p, num_trials):
    """
    Simulates random variables using binomial distribution and calculates their probabilities.

    Parameters:
    n (int): Number of trials.
    p (float): Probability of success (0 <= p <= 1).
    num_trials (int): Number of random variables to generate.

    Returns:
    dict: A dictionary with outcomes and their probabilities.
    """

    data = np.random.binomial(n, p, num_trials)

    unique_outcomes, counts = np.unique(data, return_counts=True)
    probabilities = counts / num_trials

    distribution = dict(zip(unique_outcomes, probabilities))

    return distribution

n = 10
p = 0.5
num_trials = 1000

distribution = binomial_distribution(n, p, num_trials)

print("Binomial Distribution:")
for outcome, probability in distribution.items():
    print(f"Outcome: {outcome}, Probability: {probability:.4f}")
```

#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_score(value, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    value (float): The data point
    mean (float): The mean of the dataset
    std_dev (float): The standard deviation of the dataset

    Returns:
    float: Z-score of the data point
    """
    return (value - mean) / std_dev

data_point = 185
mean = 170
std_dev = 10

z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score of the data point ({data_point}): {z_score:.2f}")

x = np.linspace(-4, 4, 500)
pdf = stats.norm.pdf(x, 0, 1)

plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.fill_between(x, pdf, where=(x <= z_score), color='skyblue', alpha=0.4, label='Cumulative Area')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Z-score Comparison with Standard Normal Distribution')
plt.legend()
plt.grid(alpha=0.6)
plt.show()
```

#15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, mu_0, alpha):
    """
    Perform a Z-test for a sample to test the null hypothesis.

    Parameters:
    sample (list or numpy array): The sample data.
    mu_0 (float): The hypothesized population mean (null hypothesis).
    alpha (float): The significance level (e.g., 0.05 for 95% confidence).

    Returns:
    None: Prints the test results.
    """

    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)

    z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis (H0) at Œ± = {alpha}.")
    else:
        print(f"Fail to reject the null hypothesis (H0) at Œ± = {alpha}.")

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
mu_0 = 55
alpha = 0.05

z_test(sample_data, mu_0, alpha)
```

#16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
    data (list or numpy array): The dataset.
    confidence (float): The confidence level (default is 95%).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    mean = np.mean(data)
    std_error = stats.sem(data)
    margin_of_error = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * std_error
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return lower_bound, upper_bound

np.random.seed(42)
data = np.random.normal(50, 10, 100)

confidence_level = 0.95
lower, upper = calculate_confidence_interval(data, confidence_level)
print(f"{confidence_level*100}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
    data (list or numpy array): The dataset.
    confidence (float): The confidence level (default is 95%).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    mean = np.mean(data)
    std_error = stats.sem(data)
    margin_of_error = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * std_error
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return mean, lower_bound, upper_bound

np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)

mean, lower, upper = calculate_confidence_interval(data)
print(f"Mean of the dataset: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

#23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Normalize the dataset using Z-score normalization.

    Parameters:
    data (numpy array): Input dataset.

    Returns:
    numpy array: Normalized dataset (Z-scores).
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=100)

normalized_data = z_score_normalization(data)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Z-score Normalized Data Distribution')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
    """
    Calculate skewness and kurtosis for a given dataset.

    Parameters:
    data (list or numpy array): The dataset.

    Returns:
    tuple: Skewness and Kurtosis values.
    """
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    return skewness, kurtosis

np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=500)

skewness, kurtosis = calculate_skewness_kurtosis(data)
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")
```
